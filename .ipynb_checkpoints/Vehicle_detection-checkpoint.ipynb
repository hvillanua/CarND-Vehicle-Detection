```python
import os
import glob
import time

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import cv2
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from scipy.ndimage.measurements import label
from moviepy.editor import VideoFileClip
import helper
```

```python
img = cv2.imread('./test_images/test1.jpg')
imshape = img.shape

roi_img = np.copy(img)
rois = [((0,380), (imshape[1],656)), ((0,380), (imshape[1],620)),
        ((0,380), (imshape[1],530))]
cv2.rectangle(roi_img, rois[0][0], rois[0][1], (255,0,0), 6)
cv2.rectangle(roi_img, rois[1][0], rois[1][1], (0,255,0), 6)
cv2.rectangle(roi_img, rois[2][0], rois[2][1], (0,0,255), 6)
helper.show_image(roi_img)
```

```python
# Read in car and non-car images
car_paths = glob.glob('./train_images/vehicles/*')
not_car_paths = glob.glob('./train_images/non-vehicles/*')
cars = []
notcars = []
for folder in car_paths:
    images = glob.glob(os.path.join(folder,'*.png'))
    cars = cars + images
    
for folder in not_car_paths:
    images = glob.glob(os.path.join(folder,'*.png'))
    notcars = notcars + images
    
cars = np.array(cars)
notcars = np.array(notcars)
print(cars.shape)
print(notcars.shape)

spatial = 16
histbin = 16
orient = 12
pix_per_cell = 16
cell_per_block = 2
cspace = 'YUV'
hog_channel = 'ALL'
#spatial = 12, histbin = 25, Acc = 0.9071

car_features = helper.extract_features(cars, cspace, (spatial, spatial), histbin,
                                      orient, pix_per_cell, cell_per_block,
                                       hog_channel, True, True, True)
notcar_features = helper.extract_features(notcars, cspace, (spatial, spatial), histbin,
                                      orient, pix_per_cell, cell_per_block,
                                       hog_channel, True, True, True)

print(len(car_features))

# Create an array stack of feature vectors
X = np.vstack((car_features, notcar_features)).astype(np.float64)                        
# Define the labels vector
y = np.hstack((np.ones(len(car_features)), np.zeros(len(notcar_features))))

# Split up data into randomized training and test sets
rand_state = np.random.randint(0, 100)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=rand_state)
```

```python
# Fit a per-column scaler
X_scaler = StandardScaler().fit(X_train)
# Apply the scaler to X
scaled_X_train = X_scaler.transform(X_train)
scaled_X_test = X_scaler.transform(X_test)

n_components = 150
t0 = time.time()
pca = PCA(n_components=n_components, whiten=True).fit(scaled_X_train)
print("PCA fit done in %0.3fs" % (time.time() - t0))
print(sum(pca.explained_variance_ratio_))
helper.plot_pca_components(pca.explained_variance_ratio_)

t0 = time.time()
X_train_pca = pca.transform(scaled_X_train)
X_test_pca = pca.transform(scaled_X_test)
print("Transformation done in %0.3fs" % (time.time() - t0))

# X_train_pca = scaled_X_train
# X_test_pca = scaled_X_test

print('Using spatial binning of:',spatial,
    'and', histbin,'histogram bins')
print('Feature vector length:', len(X_train[0]))
# Use a linear SVC 
svc = SVC(kernel='rbf', C=10, gamma=0.005)
# Check the training time for the SVC
t=time.time()
svc.fit(X_train_pca, y_train)
t2 = time.time()
print(round(t2-t, 2), 'Seconds to train SVC...')
# Check the score of the SVC
print('Test Accuracy of SVC = ', round(svc.score(X_test_pca, y_test), 4))
# Check the prediction time for a single sample
t=time.time()
n_predict = 10
print('My SVC predicts: ', svc.predict(X_test_pca[0:n_predict]))
print('For these',n_predict, 'labels: ', y_test[0:n_predict])
t2 = time.time()
print(round(t2-t, 5), 'Seconds to predict', n_predict,'labels with SVC')
```

```python
white_output = './output.mp4'
# clip1 = VideoFileClip("./test_video.mp4")
clip1 = VideoFileClip("./project_video.mp4")
```

```python
global last_heatmaps
last_heatmaps = []
global final_heatmap
final_heatmap = [0]

def pipeline(img):
    box_list = []
    global last_heatmaps
    global final_heatmap

    scales = [2, 1.5, 1]
    start_stop = [(380, 656), (380, 620), (380, 530)]

    for scale, y in zip(scales, start_stop):
        bboxes = helper.find_cars(img, y[0], y[1], scale, svc, X_scaler,
                                  pca, cspace, orient, pix_per_cell,
                                  cell_per_block, (spatial, spatial),
                                  histbin, False)
        box_list += bboxes
        
    heat = np.zeros_like(img[:,:,0]).astype(np.float)
    heat = helper.add_heat(heat, box_list)

    if len(last_heatmaps) > 10:
        final_heatmap = final_heatmap - last_heatmaps[0] + heat
        last_heatmaps = last_heatmaps[1:]
    
    last_heatmaps.append(heat)
    heatmap = np.copy(final_heatmap)
    heatmap = helper.apply_threshold(heatmap, 6)
    heatmap = np.clip(heatmap, 0, 255)

    # Find final boxes from heatmap using label function
    labels = label(heatmap)
    draw_img = helper.draw_labeled_bboxes(np.copy(img), labels)
    return draw_img
```

```python
white_clip = clip1.fl_image(pipeline)
%time white_clip.write_videofile(white_output, audio=False)
```
